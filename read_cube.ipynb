## Example on how to read a fits cube with python

```python
import pyfits as fits
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
cube = fits.open('data/NGC0309_mass_inla.fits')
cube.info()
image = np.arcsinh(cube[0].data)
```

```python
mask = (image > 0)
fig, ax = plt.subplots(figsize = (10, 10))
l1, l2 = (np.percentile(image[mask].flatten(), 10), np.percentile(image[mask].flatten(), 90))
color = ax.imshow(image,clim=(l1,l2));
fig.colorbar(color)
```
